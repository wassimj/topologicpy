```python
# This is only needed on the software development machine. If you are a user, you can skip this step.
import sys
sys.path.append("C:/Users/sarwj/OneDrive - Cardiff University/Documents/GitHub/topologicpy/src")
from topologicpy.Cell import Cell
from topologicpy.Topology import Topology

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

```

```python
def plot_and_save_mesh(vertices, faces, output_filename='mesh_plot.png'):
    # Ensure vertices and faces are numpy arrays
    vertices = np.array(vertices)
    faces = np.array(faces)

    # Create a 3D figure
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Create a collection of 3D polygons (faces)
    face_vertices = [[vertices[face] for face in face_indices] for face_indices in faces]
    poly3d = Poly3DCollection(face_vertices, facecolors='cyan', linewidths=1, edgecolors='r', alpha=.25)
    
    # Add the polygons to the plot
    ax.add_collection3d(poly3d)

    # Set limits for the axes based on the vertex positions
    ax.set_xlim([vertices[:, 0].min(), vertices[:, 0].max()])
    ax.set_ylim([vertices[:, 1].min(), vertices[:, 1].max()])
    ax.set_zlim([vertices[:, 2].min(), vertices[:, 2].max()])

    # Set axis labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Save the plot as a PNG file
    plt.savefig(output_filename)

    # Optionally, display the plot
    # plt.show()


```

```python
torus = Cell.Torus()
torus_dict = Topology.Geometry(torus)
vertices = torus_dict['vertices']
faces = torus_dict['faces']
plot_and_save_mesh(vertices, faces, output_filename="C:/Users/sarwj/OneDrive - Cardiff University/Desktop/mesh_plot.png")
```
